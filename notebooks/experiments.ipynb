# House Pricing Experiments

The following notebook contains an standalone version of the [house-pricing](..) project to perform model experiments.

### Data Ingestion

```python
from house_pricing.data_ingestion import load_house_pricing_dataset

dataset_df = load_house_pricing_dataset(
    sales_column_selection=[
        "price", 
        "bedrooms",
        "bathrooms",
        "sqft_living",
        "sqft_lot",
        "floors",
        "sqft_above",
        "sqft_basement",
        "zipcode",
    ]
)
display(dataset_df)
```

```python
dataset_full_df = load_house_pricing_dataset(
    sales_column_selection=[
        "price",
        "bedrooms",
        "bathrooms",
        "sqft_living",
        "sqft_lot",
        "floors",
        "waterfront",
        "view",
        "condition",
        "grade",
        "sqft_above",
        "sqft_basement",
        "yr_built",
        "yr_renovated",
        "zipcode",
        "lat",
        "long",
        "sqft_living15",
        "sqft_lot15",
    ]
)
display(dataset_full_df)
```

### Model Training

#### 1) KNN Model

```python
from house_pricing.model_training import train_original_model

knn_experiments = [
    {
        "name": "knn_robust",
        "df": dataset_df,
    },
    {
        "name": "knn_robust_full",
        "df": dataset_full_df,
    },
]
knn_results = {
    "train": [],
    "val": [],
    "test": [],
}
for experiment in knn_experiments:
    _, eval_metrics = train_original_model(
        df=experiment["df"],
        target_colname="price",
        robust_scaling=True,
        sampling_seed=42, 
        train_size=0.7,
        val_size=0.2,
    )
    for k in knn_results.keys():
        knn_results[k].append({
            "model": experiment["name"],
            **eval_metrics["regression_metrics"][k],
        })
```

#### 2) LightGBM Model

```python
from house_pricing.model_training import train_model

lgbm_experiments = [
    {
        "name": "lgbm",
        "df": dataset_df,
        "robust_scaling": False,
    },
    {
        "name": "lgbm_robust",
        "df": dataset_df,
        "robust_scaling": True,
    },
    {
        "name": "lgbm_full",
        "df": dataset_full_df,
        "robust_scaling": False,
    },
    {
        "name": "lgbm_robust_full",
        "df": dataset_full_df,
        "robust_scaling": True,
    },
]
lgbm_results = {
    "train": [],
    "val": [],
    "test": [],
}
for experiment in lgbm_experiments:
    _, eval_metrics, _ = train_model(
        df=experiment["df"],
        target_colname="price",
        model_hparams={
            "boosting_type": "gbdt",
            "num_leaves": 256,
            "learning_rate": 0.1,
            "n_estimators": 5000,
            "verbose": -1,
        },
        fitting_params={
            "eval_metric": ["mse"],
        },
        robust_scaling=experiment["robust_scaling"],
        sampling_seed=42, 
        train_size=0.7,
        val_size=0.2,
    )
    for k in lgbm_results.keys():
        lgbm_results[k].append({
            "model": experiment["name"],
            **eval_metrics["regression_metrics"][k],
        })
```

### Model Validation

#### 1) Experiments Comparison

```python
import pandas as pd
from IPython.display import Markdown

display(Markdown("### Train Set"))
display(pd.DataFrame.from_records(knn_results["train"] + lgbm_results["train"]))
display(Markdown("### Validation Set"))
display(pd.DataFrame.from_records(knn_results["val"] + lgbm_results["val"]))
display(Markdown("### Test Set"))
display(pd.DataFrame.from_records(knn_results["test"] + lgbm_results["test"]))
```

#### 2) Best Model

```python
model, _, _ = train_model(
    df=dataset_full_df,
    target_colname="price",
    model_hparams={
        "boosting_type": "gbdt",
        "num_leaves": 256,
        "learning_rate": 0.1,
        "n_estimators": 5000,
        "verbose": -1,
    },
    fitting_params={
        "eval_metric": ["mse"],
    },
    robust_scaling=False,
    sampling_seed=42, 
    train_size=0.7,
    val_size=0.2,
)
model
```
